```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
```

```python
Countrydatacsv = pd.read_csv(r'C:\Users\ateen\Downloads\COVID_Cases.csv')
Countrydatacsv.head()
```

```python
Countrycoviddata = pd.DataFrame(Countrydatacsv, columns = ['dateRep','year_week','cases_weekly','deaths_weekly','countriesAndTerritories','countryterritoryCode', "popData2019"])
Countrycoviddata.head()
```

```python
# for country in Countrycoviddata.columns("location"):
Indiacoviddata = Countrycoviddata[Countrycoviddata["countriesAndTerritories"] == "India"]
# pd.set_option('display.max_rows', None)
Indiacoviddata.head(10)
```

```python
LastweekdataInd =  Indiacoviddata.sort_values(by = 'year_week', ascending = False)
Last10weekdataInd = LastweekdataInd.head(10)
Last10weekdataInd
Last10weekdataIndAsc = Last10weekdataInd.sort_values(by = 'year_week')
Last10weekdataIndAsc
```

```python
plt.figure(figsize=(20, 10))
plt.plot(Last10weekdataIndAsc['dateRep'], Last10weekdataIndAsc['cases_weekly'], label = "Weekly Cases", color = "blue", marker='o', markerfacecolor='blue', markersize=2)
plt.plot(Last10weekdataIndAsc['dateRep'], Last10weekdataIndAsc["deaths_weekly"], label = "Weekly Deaths" , color = "green")
for xi,yi in zip(Last10weekdataIndAsc['dateRep'], Last10weekdataIndAsc['cases_weekly']):
    plt.text(xi,yi, str(yi))
plt.title ("COVID19 Situation in india : 26 Oct 2020 to 28 Dec. 2020 ")
plt.xticks(rotation = 45)
plt.legend()
plt.show()
```

```python
Weekdaata = Countrycoviddata[Countrycoviddata["year_week"] == "2020-50"]
Weekdaata
```

```python
TopEffectedCountry = Weekdaata.sort_values(by = "cases_weekly", ascending = False)
TopTenEffectedCountry = TopEffectedCountry.head(10)
TopTenEffectedCountry
```

```python
plt.figure(figsize=(20, 10))
xdata = TopTenEffectedCountry['countriesAndTerritories']
ydata1 = TopTenEffectedCountry['cases_weekly']
ydata2 = TopTenEffectedCountry['deaths_weekly']
plt.plot(xdata, ydata1, label = "Cases in 50th Week", color= "blue", marker = "o", markerfacecolor="black", markersize=5)
plt.plot(xdata, ydata2, label="Death in reported week", color = "red", marker = "*", markerfacecolor = "black", markersize=5)
for a,b in zip(xdata,ydata1):
    plt.text(a,b, str(b))
for c,d in zip (xdata,ydata2):
    plt.text(c,d, str(d))
# plt.figure(figsize=(108, 106), dpi=80, facecolor='w', edgecolor='k')
# plt.xlabels("Country")
plt.ylabel("Count")
plt.title("Weekly COVID19 case analysis for 50th Week i.e. 14th December 2020")
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (20,10))
plt.bar(xdata, TopTenEffectedCountry['popData2019'])
for a,b in zip (xdata, TopTenEffectedCountry['popData2019']):
    plt.text(a,b, int(b))
plt.show()
```

```python
plt.figure(figsize = (20,10))
# fig, ax1 = plt.subplots(figsize=(20,10))
color = 'tab:green'
ax1 = sns.barplot (x = xdata, y = TopTenEffectedCountry['popData2019'])
ax2 = ax1.twinx()
ax2 = sns.lineplot (x = xdata, y = TopTenEffectedCountry['cases_weekly'], color = 'blue', label = "Case Count in given week", marker = "o", markerfacecolor="black", markersize=5)
# ax3 = ax1.twinx()
ax2 = sns.lineplot (x = xdata, y = TopTenEffectedCountry['deaths_weekly'], color = 'tab:red', label = "Death Count in given week", marker = "o", markerfacecolor="black", markersize=5)

for a,b in zip(xdata,TopTenEffectedCountry['cases_weekly']):
    plt.text(a,b, str(b))
for c,d in zip(xdata,TopTenEffectedCountry['deaths_weekly']):
    plt.text(c,d, str(d))
ax1.set_ylabel("Population", fontsize=15)
ax2.set_ylabel("Case/Death Count", fontsize=15, color="blue")
ax1.set_xlabel("Countries", fontsize=20)
ax2.legend()
plt.title("Population Vs Cases Reported Vs Death Counted for 50th Week - Reported on 14 Dec 2020")
plt.show()
```

### Mortality Rate 

```python
mortrate = []
for c,p in zip(ydata2, TopTenEffectedCountry['popData2019']):
    m = c/p
    mortrate.append(m)
plt.figure(figsize = (20,10))
plt.bar(xdata, mortrate, label="Death Rate", color="brown")
plt.xticks(rotation = 90)
plt.xlabel("Countries", fontsize=20, color = "blue")
plt.ylabel("Death rate", fontsize=20, color="brown")
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.show()
```

### Interpretition from the data 
1. USA with lower population is with highest COVID cases as well as Death count
2. India with Highest population have lower COVID cases as well as death count.
3. Mortality rate in India is minimum. 
4. Mortality rate in Italy is maximum.
